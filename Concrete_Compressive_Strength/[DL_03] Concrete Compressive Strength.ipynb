# GOAL: Estimate Compressive Strength of Concrete

## About Data

### Description

```python
This intermediate level data set has 1030 rows and 9 columns.
Concrete is the most important material in civil engineering. The concrete compressive strength is a highly nonlinear function of age and ingredients
The actual concrete compressive strength (MPa) for a given mixture under a specific age (days) was determined from laboratory. Data is in raw form (not scaled).
```

### Features

```python
1. Cement (component 1)(kg in a m3 mixture) 
2. Blast Furnace Slag (component 2)(kg in a m3 mixture) 
3. Fly Ash (component 3)(kg in a m3 mixture) 
4. Water  (component 4)(kg in a m3   mixture)
5. Superplasticizer (component 5)(kg in a m3 mixture)
6. Coarse Aggregate  (component 6)(kg   in a m3 mixture)
7. Fine Aggregate (component 7)(kg in a m3 mixture)
8. Age (day)
9. *CONCRETE COMPRESSIVE STRENGTH(MPa, megapascals)
```

## Machine Learning

```python
# Splitting X and y features
X_train = train_set.drop('Concrete compressive strength(MPa, megapascals) ', axis=1)
y_train = train_set['Concrete compressive strength(MPa, megapascals) '].copy()

X_test = test_set.drop('Concrete compressive strength(MPa, megapascals) ', axis=1)
y_test = test_set['Concrete compressive strength(MPa, megapascals) '].copy()
```

```python
# Column Transformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

num_features = X_train.select_dtypes(include=['float64', 'int64']).columns

col_trans = ColumnTransformer(transformers=[
    ('num_features', StandardScaler(), num_features),
])

# Model Evaluation function and Pipeline
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score

def evaluate_model(model):
    pipeline = Pipeline(steps=[
        ('prep', col_trans),
        ('model', model),
    ])
    scores = cross_val_score(pipeline, X_train, y_train, cv=3, scoring='neg_mean_absolute_error')
    scores = np.absolute(scores)
    print('MAE for %s: %.3f (%.3f)' % (model, np.mean(scores), np.std(scores)))
```

```python
# Models
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

lin_reg = LinearRegression()
las_reg = Lasso()
ridge_reg = Ridge()
svr = SVR()
forest_reg = RandomForestRegressor()
kn_reg = KNeighborsRegressor()

models_list = [lin_reg, las_reg, ridge_reg, svr, forest_reg, kn_reg]
```

```python
# Evaluating Models
for model in models_list:
    evaluate_model(model)
```

```python
# Predicting using LinearRegression model
from sklearn.metrics import mean_squared_error

X_train_prep = col_trans.fit_transform(X_train)
X_test_prep = col_trans.fit_transform(X_test)

def predict_and_test(model):
    model.fit(X_train_prep, y_train)
    preds = model.predict(X_test_prep)
    print('MSE:', np.sqrt(mean_squared_error(preds, y_test)))

predict_and_test(lin_reg)
```

```python
# Predicting using Lasso model
predict_and_test(las_reg)
```

```python
# Predicting using Ridge model
predict_and_test(ridge_reg)
```

```python
# Predicting using SVR model
predict_and_test(svr)
```

```python
# Predicting using RandomForest model
predict_and_test(forest_reg)
```

```python
# Predicting using KNRegressor model
predict_and_test(kn_reg)
```

## Deep Learning

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Set Random Seed
tf.random.set_seed(42)

# Create Model
input_shape = [X_train_prep.shape[1]]
model = keras.Sequential([
    layers.BatchNormalization(input_shape=input_shape),
    layers.Dense(100, activation='relu'),
    layers.Dense(1),
])

# Compile Model
model.compile(
    optimizer='adam',
    loss='mae',
    metrics='mae'
)

# Fit Model
history = model.fit(X_train_prep, y_train, epochs=100)
```

```python
# Testing Model
predict_and_test(model)
```
