## Data Preprocessing
---
**Steps**
1. Remove special characters
2. Remove numbers
3. Expand contractions
4. Lowercase texts
5. Remove stopwords
6. Tokenize
7. Lemmatize

```python
# Creating functions to preprocess text data

# Importing libraries
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import re
import contractions

# Make function to preprocess text

def preprocess_text(text):
    
    # Remove special characters
    text = re.sub(r'[^a-zA-z.,!?/:;\"\'\s]', '', text) # Add step to remove XX/XX/XXXX dates
    
    # Convert to lowercase
    text = text.lower()
    
    # Expand contractions
    text = contractions.fix(text)
    
    # Tokenize text
    tokens = word_tokenize(text)
    
    # Lemmatize tokens
    tokens = [WordNetLemmatizer().lemmatize(token) for token in tokens]
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Remove whitespace
    text = ' '.join(tokens)
    text = text.strip()
    
    return text
```

```python
# Preprocess sample data
sample_data['consumer_complaint_narrative'] = sample_data['consumer_complaint_narrative'].apply(lambda text: preprocess_text(text))
```
